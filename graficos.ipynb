```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
rtt_utdt = pd.read_csv('RTTpromedioUTDT.csv')
```

```python
df_long = pd.melt(rtt_utdt, id_vars='Da', var_name='Horario', value_name='RTT')

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

plot = sns.lineplot(
    data=df_long,
    x='Da', y='RTT',
    hue='Horario',
    marker='o',
    linewidth=2.5,
    palette='magma'
)

plt.title("RTT Promedio desde UTDT por Da y Horario", fontsize=16)
plt.xlabel("Da de la Semana")
plt.ylabel("RTT (ms)")
plt.legend(title="Horario")
plt.tight_layout()

plt.savefig("RTT_promedio_UTDT", dpi=300)

plt.show()
```
